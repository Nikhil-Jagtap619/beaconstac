```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
```

```python
install_data_fb = pd.read_excel("Assignment.xlsx", sheet_name="Installs data from FB")
revenue = pd.read_excel("Assignment.xlsx", sheet_name="Revenue")
adset_mapping = pd.read_excel("Assignment.xlsx", sheet_name="Adset mapping")
adtype_mapping = pd.read_excel("Assignment.xlsx", sheet_name="Ad type mapping")
tg_to_bucket_mapping = pd.read_excel("Assignment.xlsx", sheet_name="TG to bucket mapping")
```

```python
install_data_fb.info()
```

```python
revenue.info()
```

```python
re_week = revenue.groupby("week")["revenue"].sum().reset_index()
re_week
```

```python
spend_week = install_data_fb.groupby("Week")["Amount spent"].sum().reset_index()
spend_week
```

```python
q1 = re_week.merge(spend_week, left_on="week", right_on="Week").drop(columns="Week")
q1
```

```python
q1["payback(ROI)"] = q1["revenue"]/q1["Amount spent"]*100
q1
```

## Q4

```python
q4 = revenue.groupby("fb_adset")["revenue"].sum().reset_index().sort_values("revenue", ascending=False).merge(adset_mapping, left_on="fb_adset",right_on="Ad Set Name").drop(columns="fb_adset")
```

```python
q4.sort_values("revenue", ascending=False)
```

## Q5

```python
# we already have revenue and adset names with TG
# lets find out highest amount spent
q5 = install_data_fb.groupby("Ad set name")["Amount spent"].sum().reset_index().sort_values("Amount spent", ascending=False).merge(q4, left_on="Ad set name", right_on="Ad Set Name").drop(columns="Ad Set Name")
q5
```

```python
q5["payback(ROI)"] =  q5["revenue"]/q5["Amount spent"]*100
q5.sort_values("payback(ROI)", ascending=False)
```
